```python
l=[1,2,1,3,3,1,5,6,1,2]
#op=[1,2,3,5,6] how to remove duplicates
l2=[]
for i in l:
  if i not in l2:
    l2.append(i)
print(l2)
```

```python
#find numbers which appeared for odd number of times
l=[1,2,4,3,1,1,5,5,7,5,6,8,7,8,4,8]
l2=[]
l3=[]
for i in l:
  if l.count(i)%2!=0:#and not in l2
    l2.append(i)
print(l2)
for i in l2:
  if i not in l3:
    l3.append(i)
print(l3)
```

```python
l=[3,6,2,1,5,7,8,12,11,9]
even=[]
odd=[]
for i in l:
  if i%2==0:
    even.append(i)
  else:
    odd.append(i)
even.sort()
odd.sort(reverse=True)
print(even+odd)
```

```python
l=[3,6,2,1,5,7,8,12,11,9]#with only 1 empty list
l1=[]
l.sort()
for i in l:
  if i%2==0:
    l1.insert(0,i)
  else:
    l1.append(i)
print(l1)


```

```python
#find largest number without buit-in fn
l=[3,6,2,1,5,7,8,12,11,9]
lar=l[0]
for i in l:
  if i>lar:
    lar=i
print(lar)
```

```python
#second largest number
l=[3,6,2,1,5,7,8,12,11,9]
slar=lar=l[0]
for i in l:
  if i>lar:
    slar=lar
    lar=i
  elif i>slar and i!=lar:
    slar=i
print(slar)
print(lar)
```
